```python
import re
df=dataset
# df['extracted'] = df['prediction'].str.extract(r'### Response: \[([^\]]+)\]')
def extract_last_intent(text):
    # Find all (input, intent) pairs
    matches = re.findall(r'### Input:.*?\n### Intent:\s*(\S+)', text, flags=re.DOTALL)
    return matches[-1] if matches else None

# Apply the function
df['extracted'] = df['prediction'].apply(extract_last_intent)
```

```python
# Filter only misclassified rows
mis_df = df[df['intent'] != df['extracted']]

# Group by true label and predicted label, count occurrences
confusion_pairs = (
    mis_df.groupby(['intent', 'extracted'])
    .size()
    .reset_index(name='count')
)

# For each true label, get the predicted label it was most misclassified as
most_common_confusion = (
    confusion_pairs.sort_values('count', ascending=False)
    .groupby('intent')
    .first()
    .reset_index()
)

# Rename for clarity
most_common_confusion.columns = ['True Class', 'Most Confused With', 'Count']

print(most_common_confusion)
```
